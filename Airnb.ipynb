# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import Dataset

```python
ds = pd.read_csv('compressed_data.csv')
```

```python
ds.head()
```

# Clean the Dataset

```python
ds.columns
```

```python
ds.info()
```

```python
ds.isnull().mean()*100
```

```python
ds.shape
```

```python
ds['last review'] = pd.to_datetime(ds['last review'], errors = 'coerce')
```

```python
ds.info()
```

```python
ds.fillna({'reviews per month' : 0, 'last review' : ds['last review'].min()}, inplace = True)
```

```python
ds.dropna(subset = ['NAME', 'host name'], inplace = True)
```

```python
ds.isnull().sum()
```

```python
ds.drop(['license', 'house_rules'], axis =1, inplace = True)
```

```python
ds['price'] = ds['price'].replace('[\$,]', '', regex =True).astype(float)
ds['service fee'] = ds['service fee'].replace('[\$,]', '', regex = True).astype(float)
```

```python
ds.head()
```

```python
ds.drop_duplicates(inplace =True)
```

```python
ds.describe()
```

```python
ds.isnull().sum()
```

```python
ds.fillna({'price' : ds['price'].mean()}, inplace = True)
```

```python
ds.isnull().sum()
```

```python
ds.head()
```

# Q.1 -What is the distribution of listing prices?

```python
plt.figure(figsize = (8,5))
sns.histplot(x = ds['price'], bins = 50, kde = True, color = 'red')
plt.title('Distributio of listing price')
plt.xlabel('Price in $')
plt.show()
```

The price histogram shows the even distribution of listing price ranges across different prices. 
There is a specific consideration of listing price in any particular price range
The KDE line helps us to visualise the graph more effectively

# Q2 How fiffernet room type are distributed

```python
plt.figure(figsize = (8,5))
sns.countplot(x = ds['room type'], color = 'blue', order = ds['room type'].value_counts().index)
plt.title('Room type disribution')
plt.xlabel('Count')
plt.ylabel('Room Type')
plt.show()
```

According to the distribution, more people like to book an apartment room type. 
and the second preference is a private room
Shared room and hotel room are significantly less prefered.

# Q3. How are listings distributed across different neighborhoods?

```python
plt.figure(figsize = (8,5))
sns.countplot(y = ds['neighbourhood group'], color = 'green', order = ds['neighbourhood group'].value_counts().index)
plt.title('distributedion across different neighborhoods')
plt.xlabel('Count')
plt.ylabel('neighborhoods')
plt.show()
```

# Q4. Relationship between price and room type

```python
plt.figure(figsize = (8,5))
sns.boxplot(x = 'room type', y = 'price', hue = 'room type', data = ds, palette = 'Set1')
plt.title('Price and rom type distribution')
plt.ylabel('Price in $')
plt.xlabel('Room Type')
plt.show
```
